```python
## Build dictionaries mapping row_id to cell line name, profile_id to drug name, 
## and (row_id, profile_id) to LFC value

cell_map_dic = dict(zip(df_cell['row_id'], df_cell['ccle_name']))

drug_map_dic = dict(zip(df_drug['profile_id'], df_drug['broad_id']))

LFC_mapping_dic = dict(zip(zip(df_lfc['row_id'], df_lfc['profile_id']), df_lfc['LFC']))


lfc_ls_all = []; drug_ls = []

for row_id in sorted(set(df_lfc['row_id'])):
    lfc_ls = []
    cell_ls.append(cell_map_dic[row_id])
    print(row_id)
    for profile_id in sorted(set(df_lfc['profile_id'])):
        if (row_id,profile_id) in LFC_mapping_dic:
            lfc_ls.append(LFC_mapping_dic[(row_id,profile_id)])
        else:
            lfc_ls.append(pd.NA)
        
    lfc_ls_all.append(lfc_ls)
    
df_all = pd.DataFrame(lfc_ls_all)
df_all.index = [cell_map_dic[row_id] for row_id in sorted(set(df_lfc['row_id']))]
df_all.columns = [drug_map_dic[profile_id] for profile_id in sorted(set(df_lfc['profile_id']))]

df_all
```
