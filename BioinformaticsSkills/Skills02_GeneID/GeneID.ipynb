# Convert gene IDs in batch

```python
import pandas as pd

df_gene = pd.read_excel('ZF_KRAB_list.xlsx')

print(df_gene)
```

```python
gene_list = list(df_gene['Gene symbol'])

gene_info = mg.querymany(gene_list, scopes='symbol',fields='uniprot,ensembl,entrezgene,refseq', species='9606')
```

```python
gene_info
```

```python
for info_dic in gene_info:
    #print(info_dic)
    query = info_dic['query']
    uni_id = info_dic['uniprot']['Swiss-Prot']
    ensembl_id = info_dic['ensembl']['gene']
    entrez_id = info_dic['entrezgene']
    refseq_id = info_dic['refseq']['genomic']
    
    print(query,uni_id,ensembl_id,entrez_id,refseq_id)    
```

```python
info_dic
```

```python
uni_id_ls = []

for info_dic in gene_info:
    #print(info_dic)
    query = info_dic['query']
    uni_id = info_dic['uniprot']['Swiss-Prot']
    
    if isinstance(info_dic['ensembl'], dict):
        ensembl_id = info_dic['ensembl']['gene']
    if isinstance(info_dic['ensembl'], list):
        ensembl_id = [dic['gene'] for dic in info_dic['ensembl']]
    
    entrez_id = info_dic['entrezgene']
    refseq_id = info_dic['refseq']['genomic']
    
    print(query,uni_id,ensembl_id,entrez_id,refseq_id)
    uni_id_ls.append(uni_id)
```

```python
df_gene['UniprotID'] = uni_id_ls
```
