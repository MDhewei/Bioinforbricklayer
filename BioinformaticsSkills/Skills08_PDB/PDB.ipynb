## Read PDB (Protein Data Bank) file 

```python
pdbcode = '1VJC.pdb'

p = PDBParser(PERMISSIVE=1)
PDBFile = os.path.join(os.getcwd(),'PDBDir',pdbcode) #file path to pdb
s = p.get_structure(pdbcode,PDBFile) #parse the pdb structure
chain = s[0]['A']
```

```python
## Traverse all the residues in chainA

for res in chain.get_residues():
    print(res)
```

```python
## Operations at residue level

res1 = list(chain.get_residues())[0]

print(res1.get_id())
print(res1.get_resname())

for atom in res1.get_atoms():
    print(atom)
```

```python
## Operations at atom level

atom1 = res1['CA']; atom2 = res1['CB']

print(atom1.get_parent())

print(atom1.get_name(),atom2.get_name())

print(atom1.get_coord(),atom2.get_coord())

d = atom1 - atom2  ## Calculate distance between two atoms
print(d)
```

## Calculate the angle between among three atoms

An angle is the figure formed by two lines, sharing a common endpoint (vertex).

Consider two atoms a and c, both bonded to a third atom b.

The angle formed by the three atoms a, b and c can be calculated as:

$$\varphi_{abc}=\arccos \left(\frac{\mathbf{ba} \cdot \mathbf{bc}}{|\mathbf{ba}| |\mathbf{cb}|}\right)$$
where:
$\mathbf{ba}$ is the vector (bond) conecting the a and b atoms
$\mathbf{bc}$ is the vector (bond) conecting the c and b atoms
